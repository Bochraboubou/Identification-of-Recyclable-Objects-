**Répartition en sous-dossiers par classe :** Les images appartenant à chaque classe devraient être regroupées dans des sous-dossiers distincts. Cela permet une organisation claire et facilite le traitement par lot des données.

**b. Écrire un code qui permet de mettre en forme les données d’apprentissage et de
validation selon la structure requise.**

```python

from skimage.io import imread, imsave
from skimage.transform import resize

def mise_forme_images(path_base, output_path):
    # Créer le répertoire de sortie s'il n'existe pas
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    os.chdir(path_base)
    images_list = os.listdir()
    numb_images = len(images_list)

    images = np.empty((numb_images, 64, 64, 3), dtype=np.uint8)

    k = 0
    for im in images_list:
        img = imread(im)
        img_resized = resize(img, (64, 64, 3))
        img_resized = (img_resized * 255).astype(np.uint8)

        # Construire le chemin de sortie pour chaque image
        output_filepath = os.path.join(output_path, im)
        imsave(output_filepath, img_resized)

        img = np.expand_dims(img_resized, axis=0)
        images[k, :, :, :] = img

        k += 1

    return images

```

#### **c. Concevoir un réseau CNN dont sa topologie est représentée ci-dessous. Afficher le résumé du modèle et sa topologie.**


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

# Création du modèle
model = Sequential()

# Couche 1: Convolutional layer avec input shape (64, 64, 3)
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))

# Couche 2: MaxPooling layer
model.add(MaxPooling2D((2, 2)))

# Couche 3: Convolutional layer
model.add(Conv2D(32, (3, 3), activation='relu'))

# Couche 4: MaxPooling layer
model.add(MaxPooling2D((2, 2)))

# Couche 5: Flatten layer
model.add(Flatten())

# Couche 6: Dense layer
model.add(Dense(128, activation='relu'))

# Couche 7: Dropout layer
model.add(Dropout(0.5))

# Couche 8: Dense layer
model.add(Dense(3, activation='softmax'))  # 3 classes de sortie

# Afficher la structure du modèle
model.summary()
```

#### **d. Quel est le rôle de la couche « dopout »:**
La couche Dropout dans un réseau neuronal, y compris un réseau convolutif profond, joue le rôle de régularisation en désactivant aléatoirement certains neurones pendant l'entraînement. Cela réduit le surajustement en encourageant une représentation plus robuste des caractéristiques, améliore la généralisation en obligeant le réseau à apprendre à partir de différentes combinaisons de neurones, et diminue les coadaptations entre neurones. Pendant la prédiction, la couche Dropout est désactivée pour permettre une utilisation complète du réseau.

#### **e. On s’intéresse à la première couche de convolution:**
**i. On remarque que la sortie de la couche conv2D est de taille (62, 62, 32) alors que
l’entrée est une image de taille (64, 64, 32). Expliquer cette variation de taille.**

La variation de taille de la sortie de la couche Conv2D est due à la manière dont les opérations de convolution sont effectuées. Lorsque vous appliquez une convolution à une image, la taille de la sortie dépend des paramètres tels que la taille du noyau de convolution (ici, (3, 3)) et du mode de remplissage (par défaut, qui est 'valid').

Dans ce cas, la première couche Conv2D avec un noyau de (3, 3) appliqué à une image d'entrée de taille (64, 64, 3) avec un mode de remplissage 'valid' produit une sortie de taille (62, 62, 32). La formule pour calculer la taille de sortie dans le mode 'valid' est donnée par :

La taille de sortie (\( \text{taille sortie} \)) est calculée comme suit :

\[ \text{taille sortie} = \left\lfloor \frac{\text{taille entrée} - \text{taille noyau} + 1}{\text{pas}} \right\rfloor \]





##### **ii. Taille des filtres de convolution**
La taille des filtres de convolution est déterminée par la spécification du noyau lors de la création de la couche Conv2D. Dans cet exemple, la taille du noyau est (3, 3), indiquant que chaque filtre de convolution est une matrice 3x3.

##### **iii. Nombre de filtres de convolution**
Le nombre de filtres de convolution est spécifié par l'argument filters dans la couche Conv2D. Dans cet exemple, le premier Conv2D a 32 filtres, spécifié par Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)). Cela signifie que 32 filtres de convolution différents sont appliqués à l'image d'entrée, produisant 32 canaux de sortie dans la sortie de la couche Conv2D. Chaque canal de sortie représente une carte d'activation générée par un filtre particulier.

##### **f. On s’intéresse à la couche de « pooling »**
 **i. Sorties et Entrées de la Couche de Pooling**
La couche de pooling (ici, MaxPooling2D) réduit les dimensions spatiales (largeur et hauteur) de la sortie de la couche précédente. Pour comprendre les entrées et les sorties de cette couche :

**Entrées :** Les entrées de la couche de pooling sont les sorties de la couche Conv2D qui lui est connectée. Dans votre exemple, la deuxième couche MaxPooling2D suit une couche Conv2D et prend en entrée une matrice de dimensions (62, 62, 32) provenant de la première couche Conv2D.

**Sorties :** Les sorties de la couche de pooling sont les matrices résultantes après l'application de l'opération de pooling. Dans ce cas, la sortie serait une matrice avec des dimensions réduites par rapport à l'entrée. Par exemple, si la taille de pooling est (2, 2), la sortie pourrait avoir des dimensions comme (31, 31, 32) après le pooling.

##### **ii. Taille du Filtre de Pooling**
La taille du filtre de pooling est déterminée par le paramètre de la couche de pooling. Dans l'exemple, le filtre de pooling a une taille de (2, 2), comme spécifié par MaxPooling2D((2, 2)). Cela signifie que pour chaque zone de 2x2 dans la matrice d'entrée, la valeur maximale est extraite pour former la matrice de sortie.

##### **iii. Rôle du Pooling**
Le pooling sert à deux objectifs principaux :

**Réduction de Dimension :** Il réduit la dimension spatiale de la représentation, ce qui aide à réduire le nombre de paramètres dans le modèle et à éviter le surajustement. Moins de paramètres signifient également moins de calculs, ce qui peut accélérer l'entraînement du modèle.

**Invariance aux Translations :** Le pooling rend la représentation obtenue par la couche convolutive invariante aux petites translations dans l'image d'entrée. Cela signifie que même si un objet se déplace légèrement dans l'image, les activations dans la couche de pooling ne changent pas de manière significative.

Le MaxPooling, en particulier, prend la valeur maximale dans chaque région de pooling, ce qui capture les caractéristiques les plus significatives de la région, favorisant ainsi l'invariance aux translations.

##### **g. Fonction d'activation de la Dernière Couche**
la fonction d'activation de la dernière couche est softmax. Cela est indiqué par la ligne de code suivante :

**model.add(Dense(3, activation='softmax'))**
La fonction softmax est couramment utilisée dans les tâches de classification multiclasse. Elle convertit les scores bruts en probabilités normalisées, permettant de prendre une décision finale parmi plusieurs classes.
